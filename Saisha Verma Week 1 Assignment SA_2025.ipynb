### 1. What is name of car that has the highest horsepower?  

```python
max_hp = df['horsepower'].max()
print("Car Name with Maximum horsepower : ", df[df['horsepower'] == max_hp]['name'].values[0])
```

### 2. How many cars have mpg  35?

```python
# Finding the count based on the filtered data
count = (df['mpg'] >= 35).sum()
# Printing results
print("No. of cars with mpg >= 35 : ", count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Filtering data
df1 = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
origin = df1['origin'].mode()[0]
# Printing results
print("Most Common Origin for Cars with horsepower > 100 and weight < 3000 : ", origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
# Filtering the data
df2 = df[df['origin'] == 'japan']
acc = df2['acceleration'].mean().round(2)
# Printing results
print("Mean Acceleration of Cars from Japan : ", acc)
```

### 5. Which year had the highest average mpg?

```python
# Caluclating the average mpg values by model year
df3 = df.groupby('model_year')['mpg'].mean()
high_mpg_value = df3.max()
high_mpg_year = df3.idxmax()
# Printing results
print("Highest mpg Value : ", high_mpg_value)
print("Year corresponding to the Highest mpg Vlue : ", high_mpg_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Calculating median
median_mpg = df['mpg'].median()

# Filtering data with mpg > median
df1 = df[df['mpg'] > median_mpg].copy()

# Calculating horsepower-to-weight ratio
df1['hp_per_weight'] = df1['horsepower'] / df1['weight']

# Finding maximum ratio
max_ratio = df1['hp_per_weight'].max()

# Getting cars with max ratio
cars = df1[df1['hp_per_weight'] == max_ratio]

# Printing the results
print(cars[['name', 'horsepower', 'weight', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Calculating average mpg per year per origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plotting using seaborn lineplot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
# Displaying the plot
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
# Reseting index to avoid duplicates
df = df.reset_index(drop=True) 

# Plotting the figure
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    hue_order=['japan', 'europe', 'usa'],
    alpha=0.7,
    palette='Set2'
)

plt.title('Car Weight vs Horsepower Colored by Origin and Sized by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight (lbs)', fontsize=14)
plt.legend(title='Origin', loc='upper right', fontsize=12, title_fontsize=13)
plt.grid(True, linestyle='--', alpha=0.5)
# Displaying the plot 
plt.show()
```
